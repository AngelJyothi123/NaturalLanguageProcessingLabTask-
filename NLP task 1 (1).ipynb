```python
import re
```

```python
pattern1=re.compile(r'abcd')
result1=pattern1.match("abcdefsh")
if result1:
  print("Match found:", result1.group())
```

```python
pattern2=re.compile(r'.')
result2=pattern2.match("abc,def")
if result2:
  print("Match found:", result2.group())
```

```python
pattern5=re.compile(r'\b\w')
result5=pattern5.findall("this is angel")
if result5:
  print("Match found:", result5)
```

```python
pattern3=re.compile(r'.')
result3=pattern3.match("Angel jyothi")
if result3:
  print("Match found:",result3.group())
```

```python
pattern4=re.compile(r'[aeiou]')
result4=pattern4.search("this is working abcdef")
if result4:
  print("Match found:",result4.group())
```

```python
pattern6=re.compile(r'\d{2}-\d{4}')
result6=pattern6.match('12-1234')
if result6:
  print("Match found:",result6)
```

```python
pattern7=re.compile(r'\d{2}/(\d{2})/(\d{4})')
result7=pattern7.match('12/12/2023')
if result7:
  print("Match found:",result7.group())
```

```python
pattern8=re.compile(r'cat/dog')
result8=pattern8.match('this is a cat')
if result8:
  print("Match found:",result8.group())
```
